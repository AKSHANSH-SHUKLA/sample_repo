```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV, cross_val_score, KFold


```

```python
fossil=pd.read_csv('/content/BT3041_Assignment 2_Dataset.csv')
fossil.head()
```

```python
y=fossil.age
y
```

```python
X=fossil.drop('age',axis=1)
X
```

```python
X_train1,X_test1,y_train1,y_test1=train_test_split(X,y,test_size=0.2,random_state=0)
```

#test_size=0.2

```python
X_train1.shape
```

```python
X_test1.shape
```

```python
#scaler=StandardScaler()
#X_train1_sc,X_test1_sc=scaler.fit_transform(X_train1),scaler.fit_transform(X_test1)
```

```python
lr=LinearRegression()
```

```python
y_train1.shape
```

```python
lr.fit(X_train1,y_train1)
```

```python
y_pred1=lr.predict(X_test1)
```

```python
r2_score_1=r2_score(y_test1,y_pred1)
r2_score_1
```

```python
rmse_1=np.sqrt(mean_squared_error(y_test1,y_pred1))
rmse_1
```

```python
print('R2 score is:',r2_score_1)
print('ROOT MEAN SQUARED ERROR IS:',rmse_1)
print(lr.intercept_)
print(lr.coef_)
```

#test_size=0.3

```python
X_train2,X_test2,y_train2,y_test2=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train2.shape
```

```python
X_test2.shape
```

```python
#scaler=StandardScaler()
#X_train2_sc,X_test2_sc=scaler.fit_transform(X_train2),scaler.fit_transform(X_test2)
```

```python
lr1=LinearRegression()
```

```python
lr1.fit(X_train2,y_train2)
```

```python
y_pred2=lr1.predict(X_test2)
```

```python
r2_score_2=r2_score(y_test2,y_pred2)
r2_score_2
```

```python
rmse_2= np.sqrt(mean_squared_error(y_test2,y_pred2))
rmse_2
```

```python
print('R2 score is:',r2_score_2)
print('ROOT MEAN SQUARED ERROR IS:',rmse_2)
print(lr1.intercept_)
print(lr1.coef_)
```

#REGULARIZATION MODEL

```python
from sklearn.linear_model import Ridge
```

```python
rr1=Ridge(alpha=0.1)
```

```python
rr1.fit(X_train1,y_train1)
```

```python
rr1.intercept_
```

```python
rr1.coef_
```

```python
r2_score(y_test1,rr1.predict(X_test1))
```

```python
rr1.fit(X_train2,y_train2)
```

```python
r2_score(y_test2,rr1.predict(X_test2))
```

```python
#y_pred_best_70.shape
```

```python
l = np.logspace(-3, 3, 20)
params = {'alpha': l}

ridge_cv = GridSearchCV(Ridge(), params, cv=5, scoring='neg_mean_squared_error')
ridge_cv.fit(X_train1, y_train1)

best_ridge = ridge_cv.best_estimator_

kf = KFold(n_splits=7, shuffle=True)
ridge_kf_scores = cross_val_score(Ridge(alpha=ridge_cv.best_params_['alpha']), X_train1, y_train1, cv=kf, scoring='r2')

ridge_cv_70 = GridSearchCV(Ridge(), params, cv=5, scoring='neg_mean_squared_error')
ridge_cv_70.fit(X_train2, y_train2)

best_ridge_70 = ridge_cv.best_estimator_

kf_70 = KFold(n_splits=7, shuffle=True)
ridge_kf_scores_70 = cross_val_score(Ridge(alpha=ridge_cv.best_params_['alpha']), X_train2, y_train2, cv=kf_70, scoring='r2')

y_pred_best = best_ridge.predict(X_test1)
y_pred_best_70 = best_ridge_70.predict(X_test2)


plt.figure(figsize=(10,6))
sns.scatterplot(x=y_test1, y=y_pred_best)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.title('Actual vs Predicted Fossil Age')
plt.grid(True)
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x=y_test2, y=y_pred_best_70)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.title('Actual vs Predicted Fossil Age')
plt.grid(True)
plt.show()




```
